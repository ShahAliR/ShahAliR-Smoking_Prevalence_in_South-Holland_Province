```python

south=south[~south.region_type.str.contains("Wijk")]

south=south[~south.region_type.str.contains("Gemeente")]
```

```python
def big_small(x):
    if '"Alphen aan den Rijn"' in x:
        return "B"
    elif "Delft" in x:
        return "B"
    elif "Dordrecht" in x:
        return  "B"
    elif "'s-Gravenhage" in x:
        return  "B"
    elif "Leiden" in x:
        return  "B"
    elif "Rotterdam" in x:
        return  "B"
    elif  "Westland"  in x:
        return  "B"
    elif  "Zoetermeer"  in x:
        return  "B"
    else:
        return "S"
    
    
south['big_or_small'] = list(map(big_small, south['municipality']))
```

```python
south.head(50)
```

```python
def period(x):
    if "       ." in x:
        return "NaN"
    else:
        return x
```

```python
south['normal_weight'] = list(map(period, south['normaalgewicht_8']))
```

```python
south['underweight'] = list(map(period, south['underweight']))

south['overweight'] = list(map(period, south['overweight']))
south['severe_overweight'] = list(map(period, south['severe_overweight']))
south['smoker'] = list(map(period, south['smoker']))
south['complies_with_alcohol_directive'] = list(map(period, south['complies with alcohol directive']))
south['drinker'] = list(map(period, south['drinker']))
south['heavy_drinker'] = list(map(period, south['heavy_drinker']))
south['severe_drinker'] = list(map(period, south['severe_drinker']))
south['one_or_more_long-term_conditions'] = list(map(period, south['one_or_more_long-term_conditions']))
south['restricted_because_of_health'] = list(map(period, south['restricted_because_of_health']))
south['severely_restricted-due_to_health'] = list(map(period, south['ernstigbeperktvanwegegezondheid_18']))
south['moderate_high-risk_anxiety_depression'] = list(map(period, south['moderate_high-risk_anxiety_depression']))
south['lonely'] = list(map(period, south['lonely']))
south['very_serious_lonely '] = list(map(period, south['very_serious_lonely']))
south['emotionally_lonely'] = list(map(period, south['emotionally_lonely']))
south['experienced_health_good_very_good'] = list(map(period, south['experienced_health_good_very_good']))
south['comply_with_exercise_guideline'] = list(map(period, south['voldoetaanbeweegrichtlijn_5']))
south['weekly_athletes'] = list(map(period, south['weekly_athletes']))
south['visual_restriction'] = list(map(period, south['visual_restriction']))
south['motor_restriction'] = list(map(period, south['motor_restriction']))
south['socially_lonely'] = list(map(period, south['socially_lonely']))
south['long-term_illnesses_and-limited'] = list(map(period, south['long-term_illnesses_and-limited']))
south['high-risk_anxiety_depression'] = list(map(period, south[' high-risk_anxiety_depression']))
south['hearing_restriction'] = list(map(period, south[' hearing_restriction']))
```
